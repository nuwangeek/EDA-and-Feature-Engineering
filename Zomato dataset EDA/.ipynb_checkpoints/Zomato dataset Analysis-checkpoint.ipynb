```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
## To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie chart - Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Zomato maximum record or ransaction from India
After that USA and the United Kindoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ---> Excelent
2. When ratings are between 4.0 to 4.4 ---> very good
3. When ratings are between 3.5 to 3.9 ---> good
4. When ratings are between 3.0 to 3.4 ---> Average
5. When ratings are between 2.5 to 2.9 ---> Average
6. When ratings are between 2.0 to 2.4 ---> Poor
7. When ratings are between 3.0 to 3.4 ---> Average

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1. Not rated count is very high
2. Maxximum number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
